# Importing the dependences

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Start the code

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
plt.figure(figsize=(9, 6))
plt.title('Gas comparison between USA vs. Germany vs. Canada', fontdict={'fontweight': 'bold', 'fontsize': '16'})
plt.xlabel('Years')
plt.ylabel('Gas price US$')
plt.plot(gas.Year, gas.USA, label = 'USA', c = 'blue', linestyle = '-.')
plt.plot(gas.Year, gas.Germany, label = 'Germany', c = 'green', linestyle = '--')
plt.plot(gas['Year'], gas['Canada'], label = 'Canada', c = 'red', linestyle = 'dotted')
plt.legend()
plt.xticks(gas.Year[::3])
plt.savefig('gas_prices.png', dpi = 600)
plt.show()
```
