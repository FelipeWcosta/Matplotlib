#Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head(5)
```

# Histogram

```python
binsy = [0, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]
bins = [40, 50, 60, 70, 80, 90, 99]
plt.title('Overall of FIFA players')
plt.hist(fifa.Overall, bins=bins)
plt.xticks(bins)
plt.yticks(binsy)
plt.ylabel('Number of players')
plt.xlabel('Overall')
plt.show()
```

# Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] ==  'Right'].count()[0]
labels = ['Left', 'Right']
colors = ['#3633FF', '#33E6FF']
plt.title('Preference foot of FIFA players')
plt.pie([left, right], labels = labels, colors = colors, autopct = '%.1f%%')
plt.show()
```

```python
fifa.Weight = pd.to_numeric(fifa.Weight)

plt.style.use('ggplot')

l = fifa.loc[fifa.Weight < 125].count()[0]
lm = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
m = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
mh = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
h = fifa.loc[fifa.Weight > 200].count()[0]

weights = [l, lm, m, mh, h]

labels = ['Light', 'Light Medium', 'Medium', 'Medium Heavy', 'Heavy']

explode = (0.4, 0.1, 0.1, 0.1, 0.4)

plt.title('Wiegths band of the FIFA 18 players')

plt.pie(weights, labels = labels, autopct = '%.1f%%', pctdistance = 0.8, explode = explode)

plt.show()
```

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
real_madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
```

```python
plt.style.use('default')

labels = ['FC Barcelona', 'CF Real Madrid']

plt.title('Professional football comparison')

plt.ylabel('FIFA Overall')

plt.boxplot([barcelona, real_madrid], labels = labels)

plt.show()
```
